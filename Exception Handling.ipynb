```python
try:
    print(10/0)
except ArithmeticError:
    print("Error in calculations")
except:
    print("Something else went wrong")
```

```python
try:
    print(x)
except:
    print("an Exception occured")
```

```python
try: 
    print(x)
except NameError:
    print("Variable x is not defined")
except:
    print("something alse went wrong")
```

```python
try:
    print("hello")
except:
    print("something went to wrong")
else:
    print("nothing went to wrong")
```
